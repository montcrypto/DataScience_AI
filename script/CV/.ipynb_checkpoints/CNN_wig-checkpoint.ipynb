```python
from PIL import Image
import h5py
import numpy as np
import os
import matplotlib.pyplot as plt

tdir='./data'
f=h5py.File('ILAS_2021.hdf5','w')

label=sorted([labels for labels in os.listdir(tdir) if not labels.startswith('.')])
for i in range(len(label)):
    ImgData=[]
    #g=f.create_group(os.path.join(tdir,label[i]))
    file= sorted([files for files in os.listdir(os.path.join(tdir,label[i])) if not files.startswith('.')])
    for j in range (len(file)):
        g=f.create_group(os.path.join(tdir,label[i],file[j]))
        img=Image.open(os.path.join(tdir,label[i],file[j]))
        w,h,c=np.array(img).shape
        if w<h:
            sz=int(w/2)
        else:
            sz=int(h/2)
        img_a=np.array(img)
        #
        ImgData.append(np.array(Image.fromarray(img_a[0:sz,0:sz,:]).resize((512,512))))
        ImgData.append(np.array(Image.fromarray(img_a[0:sz,sz:sz+sz,:]).resize((512,512))))
        ImgData.append(np.array(Image.fromarray(img_a[sz:sz+sz,0:sz,:]).resize((512,512))))
        ImgData.append(np.array(Image.fromarray(img_a[sz:sz+sz,sz:sz+sz,:]).resize((512,512))))
        dset=g.create_dataset('image',data=ImgData,chunks=(1,512,512,3),dtype='uint8', compression="gzip",compression_opts=9)
        f.flush()
f.flush()
f.close()
        

```

```python
# Introduction to convnets
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR) 
tf.autograph.set_verbosity(0)
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))

# example of 4 layered convnet
model = models.Sequential()

model.add(layers.Conv2D(16, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(256, (3, 3), activation='relu'))
# classifier on top of the convnet
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(target_num, activation='softmax'))

model.summary()
```
