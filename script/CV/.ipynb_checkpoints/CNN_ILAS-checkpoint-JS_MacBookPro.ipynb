```python
### data preparatiion 
imsz=128  # image size
epoch=30  # epock cycle
```

```python
from PIL import Image
import h5py
import numpy as np
import os
import matplotlib.pyplot as plt
from ILAS import *

tdir='./data'
f=h5py.File('ILAS_2021.hdf5','w')


label=sorted([labels for labels in os.listdir(tdir) if not labels.startswith('.')])
for i in range(len(label)):    
    #g=f.create_group(os.path.join(tdir,label[i]))
    file= sorted([files for files in os.listdir(os.path.join(tdir,label[i])) if not files.startswith('.')])
    for j in range (len(file)):
        ImgData=[]
        g=f.create_group(os.path.join(tdir,label[i],file[j]))
        img=Image.open(os.path.join(tdir,label[i],file[j]))
        w,h,c=np.array(img).shape
        if w<h:
            sz=int(w/2)
        else:
            sz=int(h/2)
        img_a=np.array(img)
        #
        ImgData.append(np.array(Image.fromarray(img_a[0:sz,0:sz,:]).resize((512,512))))
        ImgData.append(np.array(Image.fromarray(img_a[0:sz,sz:sz+sz,:]).resize((512,512))))
        ImgData.append(np.array(Image.fromarray(img_a[sz:sz+sz,0:sz,:]).resize((512,512))))
        ImgData.append(np.array(Image.fromarray(img_a[sz:sz+sz,sz:sz+sz,:]).resize((512,512))))
        dset=g.create_dataset('image',data=ImgData,chunks=(1,512,512,3),dtype='uint8', compression="gzip",compression_opts=9)
        f.flush()
f.flush()
f.close()
```

```python
from ILAS import *


f=h5py.File('ILAS_2021.hdf5','r')
mytrain=ILASDataRead()
mytest=ILASDataRead()
#mydata.table_sel_genus(f,'')

traingen=mytrain.ImageDataGenerator(f, class_name='', target_label='index', size=(imsz,imsz), batch_size=32, shuffle=True)
testgen=mytest.ImageDataGenerator(f, class_name='', target_label='index', size=(imsz,imsz), batch_size=32, shuffle=True)
```

```python
# Introduction to convnets
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR) 
tf.autograph.set_verbosity(0)
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))

target_num=len(label)

# example of 4 layered convnet
model = models.Sequential()

model.add(layers.Conv2D(16, (3, 3), activation='relu', input_shape=(imsz, imsz, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(256, (3, 3), activation='relu'))
# classifier on top of the convnet
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(target_num, activation='softmax'))

model.summary()
```

```python
### Training convnet on WIG 64x64 images of 119 genus
### nearly one hour by my MacbookPro

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
import time
# 処理前の時刻
t1 = time.time() 
#
res=model.fit(traingen, steps_per_epoch=8, validation_data=testgen, validation_steps=2, epochs=epoch)
# 処理後の時刻
t2 = time.time()
 
import socket
# コンピュータ名を取得
host = socket.gethostname()
print(host)  

# 経過時間を表示
elapsed_time = t2-t1
print(f"処理時間：{elapsed_time}")
```

```python
# result(res)に出力されたデータをプロットする。
# 図と計算結果のモデルを保存する。
target_class='index'
import matplotlib.pyplot as plt
%matplotlib inline

acc=res.history['accuracy']
loss=res.history['loss']
val_acc=res.history['val_accuracy']
val_loss=res.history['val_loss']
epochs=range(1,len(acc)+1)
fig, ax1= plt.subplots(figsize=(10,8))
ax2=ax1.twinx()
ax1.plot(epochs, acc, 'go', label='training accuracy')
ax2.plot(epochs, loss, 'c^', label='training loss')
ax1.plot(epochs, val_acc, 'bo', label='validation accuracy')
ax2.plot(epochs, val_loss, 'm^', label='validation loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('accuracy')
ax2.set_ylabel('loss')
ax1.legend()
ax2.legend()
plt.savefig('./history/CNN_ILAS_'+target_class+str(imsz)+'_'+str(epoch)+'epo.png')
plt.show()
# modelを保存する。
model.save('./model/CNN_ILAS_'+target_class+str(imsz)+'_'+str(epoch)+'epo.h5')
```

```python
batch_size_is=10 # sampling batch size of test image
hdf_source='./model/CNN_ILAS_index128_30epo.h5'
```

```python
from ILAS import *
h5file='ILAS_2021.hdf5'
f2 = h5py.File(h5file,'r')
test_data=ILASDataRead()
img_shape=(imsz,imsz)
target_class=''
testgen = test_data.ImageDataGenerator(f2, target_class,'index',size=img_shape,batch_size=20, shuffle=True)
target_num=len(test_data.pivot(f2,target_class,['index']))


new_model = tf.keras.models.load_model(hdf_source)
a=[]
b=[]
p=0
while p < 3:
    p=p+1
    a_ar,b_ar=testgen.__next__()
    true_classes=np.argmax(b_ar, axis=1)
    preds_ft = new_model.predict(a_ar)
    pred_classes_ft = np.argmax(preds_ft, axis=1)

    # Get the names of classes
    labellist = test_data.list_labels(f2,'','index')
    class_names=[labellist[i] for i in sorted(np.unique(labellist, return_index=True)[1])]
   

    fig, ax= plt.subplots(nrows=batch_size_is, ncols=2, figsize=(6, 25))
    plt.rcParams["font.family"] = "sans-serif"
    plt.rcParams["font.sans-serif"] = ["Hiragino sans"]
    #plt.subplots_adjust(right=0.3)
    for i in range(batch_size_is):
        best_5=np.argsort(preds_ft[i])[-5:]
        ax[i,0].imshow(a_ar[i], cmap='gray')
        ax[i,0].set_title(str(np.array(class_names)[true_classes[i]]),fontsize=8)
        ax[i,0].axis('off')
        label=np.array(class_names)[best_5]
        ax[i,1].barh(label,preds_ft[i][best_5])
        ax[i,1].set_yticklabels(label,rotation=0,ha='right', fontsize=8)       
       
    plt.savefig('./prediction/prediction_cat_00'+str(p)+'.jpg',dpi=200)
    plt.show()
```
